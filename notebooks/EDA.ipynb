# Customer Churn Exploratory Data Analysis (EDA)

This notebook explores the relationships in our customer churn dataset to inform our XGBoost model.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('../data/customer_churn_sample.csv')
df.head()
```

## Class Distribution
Let's see the overall churn rate.

```python
sns.countplot(data=df, x='churn')
plt.title('Churn Distribution')
plt.show()
```

## Features vs Churn
How does tenure and contract type affect churn?

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='churn', y='tenure')
plt.title('Tenure vs Churn')
plt.show()
```

```python
sns.countplot(data=df, x='contract', hue='churn')
plt.title('Contract Type vs Churn')
plt.show()
```
